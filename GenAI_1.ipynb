```python
get_ipython().system('pip install langchain')

from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

```python
template1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print(template1.format(topic="AI"))
```

```python
template2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print(template2.format(topic="AI", audience="beginners", tone="friendly"))
print(template2.format(topic="Python", audience="kids", tone="fun"))
print(template2.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

```python
teach = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

story = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teach.format(topic=topic))
print(interview.format(topic=topic))
print(story.format(topic=topic))
```

```python
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("user", "Explain {topic}")
])

print(chat_template.format(role="teacher", topic="Neural Networks"))
print(chat_template.format(role="interviewer", topic="Neural Networks"))
print(chat_template.format(role="motivator", topic="Neural Networks"))
```

```python
def validate_inputs(audience, tone):
    valid_aud = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_aud:
        audience = "beginner"
    if tone not in valid_tone:
        tone = "formal"

    return audience, tone
```

```python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions on {topic} for {audience} in {tone} tone",
        "storytelling": "Explain {topic} as a story for {audience} in {tone} tone"
    }

    temp = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates[style]
    )

    return temp.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
reuse_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} for {audience}"
)

inputs = [
    ("AI", "beginner"),
    ("Python", "kids"),
    ("ML", "students"),
    ("DL", "engineers"),
    ("NLP", "researchers")
]

for t, a in inputs:
    print(reuse_template.format(topic=t, audience=a))
```

```python
def pipeline(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)
    return generate_prompt(topic, audience, tone, style)

print(pipeline("AI", "beginner", "fun", "teaching"))
```
