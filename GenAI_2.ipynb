```python
import google.generativeai as genai

genai.configure(api_key="API_KEY")
model = genai.GenerativeModel('gemini-pro-latest')
```

```python
def extract(resume):
    prompt = f"""
    Extract Skills, Experience, and Tools from the resume.

    Resume:
    {resume}

    Do not assume anything not present.
    """
    return model.generate_content(prompt).text
```

```python
def match(resume_data, job):
    prompt = f"""
    Compare resume with job description.

    Resume:
    {resume_data}

    Job:
    {job}

    Give matching and missing skills.
    """
    return model.generate_content(prompt).text
```

```python
def score(match_data):
    prompt = f"""
    Based on the match, give score from 0 to 100.

    Strong: 80-100
    Average: 50-79
    Weak: below 50

    Match:
    {match_data}

    Return only number.
    """
    return model.generate_content(prompt).text
```

```python
def explain(match_data, score):
    prompt = f"""
    Explain why this score was given.

    Match:
    {match_data}

    Score:
    {score}
    """
    return model.generate_content(prompt).text
```

```python
def pipeline(resume, job):
    extracted = extract(resume)
    matched = match(extracted, job)
    sc = score(matched)
    explanation = explain(matched, sc)

    return {
        "Extracted": extracted,
        "Match": matched,
        "Score": sc,
        "Explanation": explanation
    }
```

```python
job = """
Looking for Data Scientist with:
Python, Machine Learning, Deep Learning, SQL, Data Visualization
"""

resume1 = "Python ML DL SQL Data Visualization 2 years experience"
resume2 = "Python ML SQL basic knowledge 1 year experience"
resume3 = "HTML CSS JavaScript no data science experience"
```

```python
print(pipeline(resume1, job))
print(pipeline(resume2, job))
print(pipeline(resume3, job))
```
