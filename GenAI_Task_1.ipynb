```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob

class RobustNLPEngine:
    def __init__(self):
        self.stop_words = set(stopwords.words('english'))
        self.lemmatizer = WordNetLemmatizer()

        # Simple slang dictionary (can be extended)
        self.slang_dict = {
            "u": "you",
            "r": "are",
            "btw": "by the way",
            "lol": "laugh",
            "omg": "oh my god"
        }

    # 1️⃣ Remove HTML tags
    def remove_html(self, text):
        return re.sub(r'<.*?>', '', text)

    # 2️⃣ Remove URLs
    def remove_urls(self, text):
        return re.sub(r'http\S+|www\S+', '', text)

    # 3️⃣ Remove special characters
    def remove_special_chars(self, text):
        return re.sub(r'[^a-zA-Z\s]', '', text)

    # 4️⃣ Normalize slang
    def normalize_slang(self, tokens):
        return [self.slang_dict.get(word, word) for word in tokens]

    # 5️⃣ Remove stopwords
    def remove_stopwords(self, tokens):
        return [word for word in tokens if word not in self.stop_words]

    # 6️⃣ Lemmatization
    def lemmatize(self, tokens):
        return [self.lemmatizer.lemmatize(word) for word in tokens]

    # 7️⃣ Spell correction (optional)
    def correct_spelling(self, text):
        return str(TextBlob(text).correct())

    # 🔥 Main pipeline
    def preprocess(self, text, correct_spelling=False):
        # Step 1: Cleaning
        text = text.lower()
        text = self.remove_html(text)
        text = self.remove_urls(text)
        text = self.remove_special_chars(text)

        # Step 2: Tokenization
        tokens = word_tokenize(text)

        # Step 3: Slang normalization
        tokens = self.normalize_slang(tokens)

        # Step 4: Stopword removal
        tokens = self.remove_stopwords(tokens)

        # Step 5: Lemmatization
        tokens = self.lemmatize(tokens)

        # Step 6: Join back
        processed_text = " ".join(tokens)

        # Step 7: Optional spell correction
        if correct_spelling:
            processed_text = self.correct_spelling(processed_text)

        return processed_text
```

```python
import nltk
nltk.download('punkt_tab', quiet=True)

engine = RobustNLPEngine()

text = "OMG!!! I luv this product 😍😍 Visit http://example.com now!!!"

result = engine.preprocess(text, correct_spelling=False)

print("Original:", text)
print("Processed:", result)
```
