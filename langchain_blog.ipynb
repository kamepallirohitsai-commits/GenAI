```python
LangChain is a framework used to build applications using large language models in a structured and modular way.
```

```python
from langchain.llms import OpenAI

llm = OpenAI(api_key="your-api-key")
print(llm("Explain AI in simple terms"))
```

```python
from langchain.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

print(template.format(topic="Machine Learning"))
```

```python
from langchain.chains import LLMChain

chain = LLMChain(llm=llm, prompt=template)
print(chain.run("Deep Learning"))
```

```python
from langchain.memory import ConversationBufferMemory

memory = ConversationBufferMemory()
memory.save_context({"input": "Hi"}, {"output": "Hello"})
print(memory.load_memory_variables({}))
```

```python
from langchain.agents import initialize_agent, Tool

def upper(text):
    return text.upper()

tools = [Tool(name="Uppercase", func=upper, description="Convert text to uppercase")]

agent = initialize_agent(tools, llm, agent="zero-shot-react-description")

print(agent.run("convert hello to uppercase"))
```
